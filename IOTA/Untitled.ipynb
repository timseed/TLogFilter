```python
"""
This code is to parse DXSummit's rather reluctant Spot Search for IOTA data
The data was saved as CSV. And this is how I turn it back into ISLANDS

"""



import pandas as pd
import re
```

```python
# Initial steps
#  cat *.csv | grep "," | sort | uniq > tor.csv
# Delete top 4 lines
df=pd.read_csv('tor.csv')
```

```python
df.columns=['id','de_call','dx_call','dx_info','frequency','time','dx_country','de_latitude',
            'de_longitude','dx_latitude','dx_longitude']
df.dx_call=df.dx_call.apply(lambda x: x.upper())
df.dx_info=df.dx_info.apply(lambda x: x.upper())
```

```python
df.drop(df[df['de_call'] == 'de_call'].index,inplace=True)
df.drop(df[df['dx_call'] == 'DX_CALL'].index,inplace=True)

```

```python
# We really only need 2 columns 
```

```python
df=df[['dx_call','dx_info']]
```

```python
df
```

```python
re_iota=re.compile('[ ]{0,1}([A-Z]{2}[\- ]{0,1}\d{2,3})')

def calc_island(info:str)->str:
    found=re_iota.findall(info)
    if (found):
        return found[0].strip()
    else:
        return "Unk"
    
    
def reformat(island:str)->str:
    if (len(island)==6):
        return island.replace(' ','-')
    else:
        return island[0:2]+"-"+island[-3:]
```

```python
df['island']=df.dx_info.apply(lambda x: calc_island(x))
```

```python
df.drop(df[df['island'] == 'Unk'].index,inplace=True)
df['island2']=df.island.apply(lambda x: reformat(x))
```
